# 형태소분석 

```python
# 형태소 분석기 로드
import nltk
from konlpy.tag import Okt; t = Okt()
pos_tagger = Okt()
```

```python
# Noun만 추출하는 함수
def find_text_n(x):
    temp=t.pos(x)
    
    data=[]
    
    for line in temp:
        if len(line[0])>1: #한글자 단어는 빼주는게 좋다. 
            if line[1] in ['Noun']: # 명사, 동사, 형용사 라면 
                data.append(line[0])
            
            
    return data
```

```python
# 확인
find_text_n(data['content'][1])
```

```python
# title에서의 형태소 분석 -> verb만 추출
data_pos_n=[]

for i in range(len(data)):
    
    tem=find_text_n(data['content'][i])
    for j in tem: data_pos_n.append(j)
```

```python
data_pos_n
```

```python
# 개수 확인

len(data_pos_n)
```

```python
# 중복제거

len(set(data_pos_n))
```

# 빈도수 확인

```python
# .values_counts()
#판다스에는 value_counts()라는 유용한 메서드가 있습니다. 이 메서드를 사용하면 각각의 고유 값(unique value)의 개수를 반환합니다.

word_list = pd.Series(data_pos_n)
result = word_list.value_counts()
result
```
