# 감성사전 로드

```python
# 사전 로드 

e_word2=pd.read_excel('D:\\toy\\data\\keywords_v0.7.xlsx')
```

```python
# word 컬럼을 리스트로 변환

e_w=[]
for k in e_word2['keyword']:
    e_w.append(k)

print(e_w)
```

```python
# num 컬럼을 리스트로 변환

e_n=[]
for k in e_word2['score']:
    e_n.append(k)

print(e_n)
```

```python
# e_w, w_n을 사전으로 변환

e_word3=dict(zip(e_w,e_n))
```

# content 토큰들과 감성사전 매칭

```python
# data_no3와 긍/부정 사전 매칭

e_word=[]
for i in data_no3:
    e_word_1=[]
    for j in i:
        if j in e_word3.keys():
            c=[j,e_word3.get(j)]
            e_word_1.append(c)
        else:
            c=[j,'None']
            e_word_1.append(c)
    e_word.append(e_word_1)

e_word[:8]
```

# content  문서별 긍/부정 점수 계산

```python
# 각 문서 별 긍/부정 점수 계산
doc_num=[]
for i in data_all2['c_n_pos_pn']:
    s_num_3=len(i)
    if i == [] or i == ['']:
        t_s=0
    else:
        s_num_1=0
        s_num_2=0
        for j in range(len(i)):
            if i[j][1] == 1.5:
                s_num_1 += 1
            elif i[j][1] == -1.5:
                s_num_2 += 1
        t_s = round((s_num_1- s_num_2) /s_num_3,2)
    doc_num.append(t_s)
```
