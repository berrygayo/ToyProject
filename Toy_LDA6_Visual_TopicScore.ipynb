# 전체 토픽에 대한 df 데이터 셋

```python
# 토픽별 60일 전 평균을 구함

topic_all=[]
for i in range(20):
    a=D1_202['topic{}'.format(i)].rolling(window=60).mean()  # 토픽별 이전 60일 간의 평균을 구함
    topic_all.append(a)
```

```python
# 확인

topic_all
```

```python
# 리스트로 변환

for i in range(len(topic_all)):
    topic_all[i]=topic_all[i].tolist()
```

```python
# 개수 확인

len(topic_all)
```

```python
# Df 데이터 셋 생성

topic_all_re=[]
for k in range(20):
    topic_all_re2=[]
    for j in range(len(D1_202)):
        if j > 60:   # 인덱스 번호가 60보다 크면
            topic_all_re2.append(D1_202['topic{}'.format(k)][i]-topic_all[k][j])  # 원본의 토픽별 일자별 원소에서 토픽별 일자별 평균을 뺌
        else:
            topic_all_re2.append('NaN') # 60이하면 NaN 입력
    topic_all_re.append(topic_all_re2)
```

```python
# 데이터프레임으로 변환

topic_all_re2=pd.DataFrame(topic_all_re)
```
