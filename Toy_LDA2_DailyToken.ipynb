```python
import pandas as pd
import datetime as dt

# 뉴스 파일 로드

data_all3=pd.read_excel('D:\\toy\\data\\online_news_i3.xlsx')
```

```python
# 확인

data_all3.columns
```

# created_date에서 년월일만 추출하기

```python
# created_date 컬럼을 datetime으로 변환

data_all3['created_date']=pd.to_datetime(data_all3['created_date'])
```

```python
# 확인

data_all3.info()
```

```python
# ymd 컬럼 생성

data_all3['ymd']=data_all3['created_date'].dt.strftime('%Y-%m-%d')
```

```python
# 확인

data_all3.info()
data_all3['ymd']
```

# 일자별 키워드 모음 생성

```python
# T1 변수 생성

T2=data_all3
```

```python
# 컬럼 삭제

del T2['Index']
del T2['created_date']
```

```python
# 컬럼 순서 변경

T2=T2[['ymd','sw_n']]
```

```python
# 확인

T2
```

```python
T2['ymd'][T2['ymd']=='2016-01-08']
```

```python
# 일별 sw_n 통합

T3=pd.DataFrame(T2).groupby('ymd').sum()
```

```python
len(T3)
```

```python
T4=T3.values.tolist()
```

```python
len(T4[0][0])
```

```python
# pos_c 리스트안의 문자형 분리 & 특수문자 제거

t_sp=[]
for j in T4:
    for i in range(len(j)):
        t_sp.append(j[i].split(','))
    
t_sp[0]
```

```python
t_st=[]
for i in range(len(t_sp)):
    t_st2=[]
    for j in t_sp[i]:
        t_st2.append(j.lstrip('[').rstrip('"]').strip(" ' "))
    t_st.append(t_st2)

t_st[0][1]
```

```python
t_st[0]
```

```python
t_st2=[]
for i in t_st:
    a=','.join(i)
    a2=a.lstrip("['").rstrip("']").replace("['","").replace("']","")
    t_st2.append(a2)
```

```python
t_st2[1]
```

```python
T1=pd.DataFrame(t_st2)
```
