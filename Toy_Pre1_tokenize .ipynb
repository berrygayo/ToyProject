```python
# Verb만 추출하는 함수

def find_text_v(x):
    temp=t.pos(x)
    
    data=[]
    
    for line in temp:
        if len(line[0])>1: #한글자 단어는 빼주는게 좋다. 
            if line[1] in ['Verb']: # 명사, 동사, 형용사 라면 
                data.append(line[0])
            
            
    return data
```

```python
# Adjective만 추출하는 함수

def find_text_a(x):
    temp=t.pos(x)
    
    data=[]
    
    for line in temp:
        if len(line[0])>1: #한글자 단어는 빼주는게 좋다. 
            if line[1] in ['Adjective']: # 명사, 동사, 형용사 라면 
                data.append(line[0])
            
            
    return data
```

```python
# 확인

find_text_v(data['title'][1])
```

```python
# title에서의 형태소 분석 -> verb만 추출

data_pos_v=[]

for i in range(len(data)):
    
    tem=find_text_v(data['title'][i])
    for list in tem: data_pos_v.append(list)
```
